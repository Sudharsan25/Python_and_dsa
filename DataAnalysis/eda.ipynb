```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
# check for missing values
df.isnull().sum()
```

```python
## filling missing values with 0
df.fillna(0, inplace=True)
# check again for missing values    
df.isnull().sum()
```

```python
# renaming columns

df.rename(columns={'ORDERNUMBER': "Order No.",}, inplace=True)
```

```python
df.head()
```

```python
# Chaning the data type of a column

df['QTR_ID'] = df['QTR_ID'].astype(float)
df['QTR_ID'].dtype
```

```python
df.head()
```

```python
# chaning column values

df['SALES'] = df['SALES'].apply(lambda x: x * 1000)
df.head()  
```

```python
# data aggregation and grouping

df_grouped = df.groupby('COUNTRY')['SALES'].mean()
df_grouped
```

```python
df_grouped = df.groupby('COUNTRY')['PRICEEACH'].sum()
df_grouped
```

```python
# aggregate multiple functions

df_grouped = df.groupby('COUNTRY').agg({'SALES': ['mean', 'sum'], 'QUANTITYORDERED': ['mean', 'sum']})
df_grouped
```
